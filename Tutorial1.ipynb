# About torch.Tensor function
함수 torch.Tensor는 원하는 텐서에 메모리를 할당하지만 메모리에 이미 있는 모든 값을 재사용합니다. 초기화 중에 텐서에 값을 직접 할당하려면 다음과 같은 여러 가지 방법이 있습니다:

torch.zeros: 0으로 채워진 텐서를 작성합니다

torch.ones: 텐서를 작성합니다. 텐서를 1개씩 채웁니다

torch.rand: 0과 1 사이에서 균일하게 샘플링된 랜덤 값으로 텐서를 만듭니다

torch.randn: 평균이 0이고 분산이 1인 정규 분포에서 추출한 랜덤 값으로 텐서를 만듭니다

torch.arange: N, N+1, N+2, ... M 값을 포함하는 텐서를 작성합니다.

Torch.Tensor(input list): 제공하는 리스트 요소에서 텐서를 작성합니다

```python
#Create a tensor from a (nested) list 내포된 리스트에서 텐서 만들기 원하는 요소를 넣어서 텐서를 만든다 ?
x = Tensor([[1,2],[3,4]])
print(x)
```

```python
#Create a tensor with random values between 0 and 1 with the shape [2,3,4]
x= torch.rand(2,3,4)
print(x)
```

numpy(x.shape)에서와 동일한 방법으로 또는 .size 방법을 사용하여 텐서의 모양을 얻을 수 있습니다:

```python
shape = x.shape
print("Shape:", x.shape)
print("Shape:", shape)

size = x.size()
print("Size:", size)

dim1, dim2, dim3 = x.size()
print("Size:", dim1, dim2, dim3)

#결론 = .shape .size() 둘 다 똑같이 텐서의 크기를 알 수 있음
```

텐서에서 Numpy로, Numpy에서 Tensor로

텐서는 numpy 배열로 변환되고 numpy 배열은 다시 텐서로 변환될 수 있습니다. numpy 배열을 텐서로 변환하려면 torch.from_numpy 사용

```python
np_arr = np.array([[1,2],[3,4]])
tensor = torch.from_numpy(np_arr)

print("Numpy array:", np_arr)
print("PyTorch tensor:",tensor)
```

PyTorch 텐서를 numpy 배열로 다시 변환하려면 텐서에서 .numpy() 함수를 사용할 수 있습니다:

```python
tensor = torch.arange(4) #0부터 시작해서 4개의 원소를 가진 텐서를 생성
np_arr = tensor.numpy() #생성한 텐서를 넘파이 배열로 바꿈

print("PyTorch tensor:", tensor)
print("Numpy array:", np_arr)
```

텐서를 numpy로 변환하려면 텐서가 GPU가 아닌 CPU에 있어야 합니다(이후 섹션의 GPU 지원에 대한 자세한 내용). GPU에 텐서가 있을 경우 텐서에 .cpu()를 미리 호출해야 합니다. 따라서 np_arr = tensor.cpu().numpy()와 같은 선을 얻을 수 있습니다.


#Operation

numpy에 존재하는 대부분의 연산은 PyTorch에도 존재합니다. 작업의 전체 목록은 PyTorch 설명서에서 찾을 수 있지만 여기서는 가장 중요한 작업을 검토합니다.

가장 간단한 작업은 두 개의 텐서를 추가하는 것입니다:

```python
x1 = torch.rand(2,3)
x2 = torch.rand(2,3)
y = x1 + x2

print("X1",x1)
print("X2",x2)
print("Y",y)
```

x1 + x2를 호출하면 두 입력의 합을 포함하는 새 텐서가 생성됩니다. 그러나 텐서의 메모리에 직접 적용되는 인플레이스 연산도 사용할 수 있습니다. 따라서 작업 전에 x2 값에 다시 액세스할 기회 없이 x2 값을 변경합니다. 예는 다음과 같습니다:

```python
x1 = torch.rand(2,3)
x2 = torch.rand(2,3)
print("X1(before)",x1)
print("X2(before)",x2)

x2.add_(x1)  # x2라는 텐서에 x1이라는 텐서를 더해준 것이 바로 x2로 업데이트 됨
print("X1 (after)", x1)
print("X2 (after)", x2)
```

인플레이스 작업은 일반적으로 밑줄 포스트픽스(예: torch.add 대신 torch.add_)로 표시됩니다.

또 다른 일반적인 작업은 텐서의 모양을 바꾸는 것을 목표로 합니다. 크기(2,3)의 텐서는 동일한 수의 요소(예: 크기(6)의 텐서 또는 (3,2)의 텐서)로 다른 모양으로 재구성할 수 있습니다. PyTorch에서 이 작업을 view라고 합니다:

요소 수는 동일하지만 배치를 다르게 하는 작업 => view

```python
x = torch.arange(6)
print("X",x)
```

```python
x = x.view(2,3)
print("X",x)
```

```python
x = x.permute(1,0) # Swapping dimension 0 and 1 행과 열 바꾸기 = 전치
print("X",x)
```

일반적으로 사용되는 다른 연산에는 신경망에 필수적인 행렬 곱셈이 포함됩니다. 학습된 가중치 행렬 W를 사용하여 변환되는 입력 벡터 X가 있는 경우가 많습니다. 행렬 곱셈을 수행하는 방법과 함수는 여러 가지가 있으며, 그 중 일부는 다음과 같습니다:

torch.matmul: 두 개의 텐서에 대해 행렬 곱을 수행합니다. 여기서 특정 동작은 치수에 따라 달라집니다. 두 입력이 모두 행렬(2차원 텐서)인 경우 표준 행렬 곱을 수행합니다. 고차원 입력의 경우 이 기능은 브로드캐스트를 지원합니다. numpy와 유사한 @b로도 쓸 수 있습니다.

torch.mm : 두 개의 매트릭스 상에서 매트릭스 제품을 수행하지만 브로드캐스트는 지원하지 않습니다

torch.bmm: 지원 배치 치수로 행렬 제품을 수행합니다. 첫 번째 텐서 T가 모양(bxnxm)이고 두 번째 텐서 R(bxmxp)이라면, 출력 O는 모양(bxmxp)이며, T와 R: O_i = T_i @R_i의 하위 행렬의 b 행렬 곱셈을 수행하여 계산되었습니다

torch.einsum: 아인슈타인 합계 규칙을 사용하여 행렬 곱셈 등(즉, 곱셈)을 수행합니다. 아인슈타인의 합에 대한 설명은 과제 1에서 찾을 수 있습니다.

보통 torch.matmul 또는 torch.bmm를 사용합니다. 아래의 torch.matmul을 사용하여 행렬 곱셈을 시도할 수 있습니다.

```python
x = torch.arange(6)
print(x)
x = x.view(2,3)
print("\n X",x)
```

```python
W = torch.arange(9).view(3,3) # 여러 작업을 한 줄로 쌓을 수도 있습니다
print("W",W)
```

```python
h = torch.matmul(x,W) #두 텐서의 행렬곱
print("h",h)
```

#Indexing

우리는 종종 텐서의 일부를 선택해야 하는 상황이 발생합니다. 인덱싱은 numpy에서와 마찬가지로 작동하므로 다음과 같이 시도해 보겠습니다:

```python
x = torch.arange(12).view(3,4)
print("X",x)

print(x[:,1]) # 두번째 열
print(x[0]) # 첫번째 행
print(x[:2,-1]) # 처음부터 2개의 행, 마지막 열
print(x[1:3,:]) # 중간 2개의 행 = 1(두번째행)부터 3(네번째행)전까지
```

#Dynamic Computation Graph and Backpropagation
/동적 계산 그래프 및 역전파


딥러닝 프로젝트에서 PyTorch를 사용하는 주요 이유 중 하나는 우리가 정의한 함수의 그레이디언트/도함수를 자동으로 얻을 수 있기 때문입니다. 우리는 주로 신경망을 구현하기 위해 파이토치를 사용할 것이며, 그것들은 단지 화려한 기능일 뿐입니다. 만약 우리가 배우고자 하는 함수에 가중치 행렬을 사용한다면, 그것들은 매개변수 또는 단순히 가중치라고 불립니다.

우리의 신경망이 단일 스칼라 값을 출력한다면, 우리는 도함수를 취하는 것에 대해 이야기할 것이지만, 당신은 우리가 여러 개의 출력 변수("값")를 갖는 것을 꽤 자주 보게 될 것입니다. 그런 경우 우리는 기울기에 대해 이야기합니다. 좀 더 일반적인 용어입니다.

입력 X가 주어지면 일반적으로 가중치 행렬을 사용한 행렬 곱셈과 소위 바이어스 벡터를 사용한 추가를 통해 입력을 조작하여 함수를 정의합니다. 입력을 조작함에 따라 자동으로 계산 그래프가 만들어집니다. 이 그래프는 입력에서 출력에 도달하는 방법을 보여줍니다. PyTorch는 실행에 따라 정의되는 프레임워크입니다. 이것은 우리가 우리의 조작을 할 수 있다는 것을 의미합니다. PyTorch는 우리를 위해 그 그래프를 추적할 것입니다. 따라서, 우리는 그 과정에서 동적 계산 그래프를 만듭니다.

자, 요약하자면, 우리가 해야 할 일은 출력을 계산하는 것뿐입니다. 그리고 나서 우리는 PyTorch에게 자동으로 기울기를 얻을 수 있도록 요청할 수 있습니다.

참고: 경사도가 필요한 이유는 무엇입니까? 입력 벡터 X에 대한 특정 출력 y를 계산해야 하는 함수인 신경망을 정의했다고 생각해 보십시오. 그런 다음 네트워크가 얼마나 잘못되었는지, 입력 X에서 출력 y를 예측하는 데 얼마나 나쁜지 알려주는 오류 측정을 정의합니다. 이 오차 측정을 기반으로 기울기를 사용하여 가중치 W은(는) 출력을 담당하므로 다음 번에 네트워크에 입력 X를 표시할 때 출력이 원하는 것에 더 가까워질 것입니다.

우리가 해야 할 첫 번째 일은 기울기가 필요한 텐서를 지정하는 것입니다. 기본적으로 텐서를 만들 때는 gradients가 필요하지 않습니다.

```python
x= torch.ones((3,))
print(x.requires_grad)
```

requires_grad_() 함수를 사용하여 기존 텐서에 대해 이 값을 변경할 수 있습니다(이것이 in-place 연산임을 나타내는 밑줄). 또는 텐서를 만들 때 requires_grad=True 인수를 위에서 본 대부분의 이니셜라이저에 전달할 수 있습니다.

```python
x.requires_grad_(True)
print(x.requires_grad)
```

```python
#device 지정

device = torch.device("cuda") if torch.cuda.is_available else torch.device("cpu") # 문법
print("Device", device)
```

GPU가 있는 경우 텐서 옆에 속성 device='cuda:0'이 인쇄되어 있는 것을 볼 수 있습니다. cuda 옆에 있는 0은 이 장치가 컴퓨터의 0번째 GPU 장치임을 나타냅니다. PyTorch는 다중 GPU 시스템도 지원하지만, 교육할 대규모 네트워크가 있는 경우에만 필요합니다(관심 있는 경우 PyTorch 설명서 참조). 또한 CPU의 대규모 매트릭스 곱셈의 런타임을 GPU의 연산과 비교할 수 있습니다:

```python
x = torch.randn(5000,5000)

#CPU version
start_time = time.time()
_=torch.matmul(x,x)
end_time = time.time()
print(f"CPU time: {(end_time - start_time):6.5f}s")

#GPU version
if torch.cuda.is_available():
  x = x.to(device)
  #CUDA is asynchronous, so we need to use different timing functions #CUDA는 비동기식이므로 다른 타이밍 기능을 사용해야 합니다
  start = torch.cuda.Event(enable_timing=True)
  end = torch.cuda.Event(enable_timing=True)
  start.record()
  _=torch.matmul(x,x)
  end.record()
  torch.cuda.synchronize() # GPU에서 모든 작업이 완료될 때까지 기다림
  print(f"GPU time: {0.001 * start.elapsed_time(end):6.5f}s") # 밀리초에서 초까지
```

작업 크기와 시스템의 CPU/GPU에 따라 이 작업의 속도가 50x 이상이 될 수 있습니다. 산술 연산은 신경망에서 매우 일반적이기 때문에 GPU에서 NN을 훈련시키는 것의 큰 이점을 이미 볼 수 있습니다. 여기서는 여러 번 실행하지 않았기 때문에 시간 견적이 상대적으로 시끄러울 수 있습니다. 이것을 연장하는 것은 자유지만, 달리는 데도 더 오래 걸립니다.

# Simple classfier
이제 Torch.nn 패키지에 미리 정의된 모듈을 사용하여 자체적인 작은 신경망을 정의할 수 있습니다. 입력 계층이 있는 최소 네트워크, 활성화 기능이 있는 숨겨진 계층 하나, 출력 계층이 있는 네트워크를 사용할 것입니다. 즉, 네트워크는 다음과 같이 표시되어야 합니다:

image.png

입력 뉴런은 파란색으로 표시되며, 이는 데이터 포인트의 좌표 x_1 및 x_2를 나타냅니다. tanh actication를 포함한 숨겨진 뉴런은 흰색으로 표시되고 출력 뉴런은 빨간색으로 표시됩니다. PyTorch에서 이를 다음과 같이 정의할 수 있습니다:

```python
class SimpleClassifier(nn.Module):
  def __init__(self, num_inputs, num_hidden, num_outputs):
    super().__init__()
    # 네트워크 구축에 필요한 모듈 초기화
    self.linear1 = nn.Linear(num_inputs, num_hidden)
    self.act_fn = nn.Tanh()
    self.linear2 = nn.Linear(num_hidden, num_outputs)

  def forward(self,x):
    # 모형 계산을 수행하여 예측 결정
    x = self.linear1(x)
    x = self.act_fn(x)
    x = self.linear2(x)
    return x
```

이 노트북의 예를 들어, 우리는 두 개의 입력 뉴런과 네 개의 숨겨진 뉴런이 있는 작은 신경망을 사용할 것입니다. 이진 분류를 수행할 때 단일 출력 뉴런을 사용합니다. Sigmoid는 아직 출력에 적용되지 않습니다. 이는 다른 함수, 특히 손실이 시그모이드 출력 대신 원래 출력을 계산하는 데 더 효율적이고 정확하기 때문입니다. 자세한 이유는 추후에 논의하도록 하겠습니다.

```python
model = SimpleClassifier(num_inputs=2, num_hidden=4, num_outputs=1)
# 모듈을 인쇄하면 모든 하위 모듈이 표시됨
print(model)
```

모델을 인쇄하면 모델에 포함된 모든 하위 모듈이 나열됩니다. 모듈의 매개 변수는 parameters() 함수를 사용하거나 named_parameters()를 사용하여 각 매개 변수 개체의 이름을 가져올 수 있습니다. 소규모 신경망의 경우 다음과 같은 매개 변수가 있습니다:

```python
for name, param in model.named_parameters():
  print (f"Parameter {name}, shape {param.shape}")
```

각 선형 레이어에는 모양[output, input]의 가중치 행렬과 모양[output]의 바이어스가 있습니다. tanh 활성화 기능에는 파라미터가 없습니다. 매개 변수는 nn.Module에 대해서만 등록됩니다.직접적인 개체 속성인 모듈 개체(예: self.a = ...)... 모듈 목록을 정의하면 모듈의 매개 변수가 외부 모듈에 등록되지 않으므로 모듈을 최적화할 때 몇 가지 문제가 발생할 수 있습니다. nn.ModuleList, nn.ModuleDict, nn.Sequential과 같은 대안들이 있습니다. 이들은 당신이 서로 다른 모듈의 데이터 구조를 가지게 해줌. 나중에 몇 가지 튜토리얼에서 사용하고 거기서 설명하겠습니다.

# The dataset class
데이터 세트 클래스는 데이터 세트의 기본 기능을 자연스럽게 요약합니다. PyTorch에서 데이터 세트를 정의하기 위해 _ _ getitem _ _(원래는 언더바사이에 띄어쓰기가 없다)과 _ _len _ _의 두 가지 함수를 지정하기만 하면 됩니다.

 get-item 함수는 데이터 세트의 i번째 데이터 포인트를 반환해야 하는 반면, len 함수는 데이터 세트의 크기를 반환합니다. XOR 데이터 세트의 경우 다음과 같이 데이터 세트 클래스를 정의할 수 있습니다:

```python
class XORDataset(data. Dataset):
  def __init__(self, size, std=0.1):
    """
    Inputs:
          size - 생성할 데이터 점의 수
          std - 노이즈의 표준 편차(generate_continuous_xor 함수 참조)
   """
    super().__init__()
    self.size = size
    self.std = std
    self.generate_continuous_xor()

  def generate_continuous_xor(self):
    # XOR 데이터 집합의 각 데이터 점에는 0 또는 1일 수 있는 두 변수 x와 y가 있습니다
    # 레이블은 XOR 조합입니다. 즉, x 또는 y만 1이고 y만 0인 경우 1입니다.
    # x=y인 경우 레이블은 0입니다.
    data = torch.randint(low = 0, high = 2, size = (self.size, 2), dtype = torch.float32)
    label = (data.sum(dim=1) == 1).to(torch.long)
    # 이를 조금 더 어렵게 만들기 위해 데이터 포인트에 가우스 노이즈를 약간 추가합니다.
    data += self.std * torch.randn(data.shape)

    self.data = data
    self.label = label

  def __len__(self):
    # 데이터 포인트 수. 또는 self.data.shape[0] 또는 self.label.모양[0]
    return self.size

  def __getitem__(self, idx):
    # 데이터 집합의 idx번째 데이터 지점 반환
    # 반환할 항목(데이터 포인트 및 레이블)이 여러 개인 경우 튜플로 반환할 수 있습니다
    data_point = self.data[idx]
    data_label = self.label[idx]
    return data_point, data_label
```

이러한 데이터 세트를 생성하여 검사해 보겠습니다:

```python
dataset = XORDataset(size = 200)
print("Size of dataset:", len(dataset))
print("Data point 0:", dataset[0])
```

```python
def visualize_samples(data, label):
  if isinstance(data, Tensor):
    data = data.cpu().numpy()
  if isinstance(label, Tensor):
    label = label.cpu().numpy()
  data_0 = data[label == 0]
  data_1 = data[label == 1]

  plt.figure(figsize=(4,4))
  plt.scatter(data_0[:,0],data_0[:,1], edgecolor="#333", label="Class 0")
  plt.scatter(data_1[:,0],data_1[:,1], edgecolor="#333", label="Class 1")
  plt.title("Dataset samples")
  plt.ylabel(r"$x_2$")
  plt.xlabel(r"$x_1$")
  plt.legend
```

# The data loader class

클래스 torch.utils.data.DataLoader는 자동 배치, 다중 프로세스 데이터 로드 및 더 많은 기능을 지원하는 데이터 세트에 대한 Pythonitable을 나타냅니다. 데이터 로더는 _ _ getitem _ _ 함수를 사용하여 데이터 세트와 통신하고 출력을 첫 번째 차원에 텐서로 쌓아 배치를 구성합니다. 데이터 세트 클래스와 달리 일반적으로 데이터 로더 클래스를 정의할 필요는 없지만 데이터 세트를 입력으로 하여 개체를 만들 수 있습니다. 또한 다음과 같은 입력 인수를 사용하여 데이터 로더를 구성할 수 있습니다(선택 항목만 해당, 여기에서 전체 목록 참조):

batch_size: 배치당 쌓을 샘플 수

shuffle: True이면 데이터가 랜덤 순서로 반환됩니다. 이것은 확률성을 도입하기 위한 훈련 중에 중요합니다.

num_workers: 데이터 로드에 사용할 하위 프로세스 수입니다. 기본값인 0은 데이터가 메인 프로세스에 로드됨을 의미하며, 이는 데이터 포인트 로드에 상당한 시간이 걸리는 데이터 세트의 교육 속도를 늦출 수 있습니다(예: 큰 이미지). 이러한 작업자는 더 많은 작업자가 권장되지만 윈도우즈 시스템에 문제가 발생할 수 있습니다. 우리와 같은 소규모 데이터 세트의 경우 일반적으로 0명의 작업자가 더 빠릅니다.

pin_memory: True이면 데이터 로더는 텐서를 반환하기 전에 CUDA 고정 메모리에 복사합니다. 이를 통해 GPU의 대용량 데이터 포인트를 위한 시간을 절약할 수 있습니다. 일반적으로 교육 세트에 사용하는 것이 좋지만 GPU의 메모리를 절약하기 위해 검증 및 테스트에 반드시 사용할 필요는 없습니다.

drop_last: True인 경우 지정된 배치 크기보다 작은 경우 마지막 배치가 삭제됩니다. 이 문제는 데이터 세트 크기가 배치 크기의 배수가 아닌 경우에 발생합니다. 일관된 배치 크기를 유지하기 위해 교육 중에만 유용합니다.

# Optimization

모델과 데이터 세트를 정의한 후에는 모델의 최적화를 준비해야 합니다. 교육 중에 다음 단계를 수행합니다:

1. 데이터 로더에서 배치 가져오기

2. 배치에 대한 모형의 예측값 가져오기

3. 예측과 레이블 간의 차이를 기반으로 손실 계산

4. 역전파: 손실과 관련된 모든 매개변수에 대한 기울기 계산

5. gradiant 방향으로 모델의 매개변수를 업데이트

PyTorch에서 1단계, 2단계 및 4단계를 수행하는 방법을 살펴보았습니다. 이제 3단계와 5단계를 살펴보겠습니다.

#Stochastic Gradient Descent

매개 변수를 업데이트하기 위해 PyTorch는 가장 널리 사용되는 최적화 도구가 구현된 패키지 torch.optim을 제공합니다. 구체적인 최적화 도구와 그 차이점에 대해서는 나중에 설명하겠지만, 현재로서는 가장 간단한 것인 torch.optim.SGD을 사용할 것입니다. 확률적 그레이디언트 강하는 그레이디언트에 학습 속도라고 하는 작은 상수를 곱하고 매개 변수에서 이러한 상수를 빼서 매개 변수를 업데이트합니다(따라서 손실을 최소화함). 따라서, 우리는 손실을 최소화하는 방향으로 천천히 나아갑니다. 소규모 네트워크에 대한 학습 속도의 양호한 기본값은 0.1입니다.

```python
# 최적화 도구에 대한 입력은 모델의 매개 변수입니다. model.parameters()
optimizer = torch.optim.SGD(model.parameters(), lr = 0.1)
```

옵티마이저는 optimizer.step()과 optimizer.zero_grad() 두 가지 유용한 함수를 제공합니다. step 함수는 위에서 설명한 대로 기울기를 기반으로 매개변수를 업데이트합니다. optimizer.zero_grad() 함수는 모든 매개변수의 기울기를 0으로 설정합니다. 이 함수는 처음에는 덜 중요해 보일 수 있지만, 역전파를 수행하기 전에 필수적인 사전 단계입니다. 이전 배치에서 매개변수 기울기가 0이 아닌 경우 손실에 대한 backward 함수를 호출하면 새로운 기울기가 이전 것에 추가되어 덮어쓰지 않습니다. 이것은 매개변수가 계산 그래프에서 여러 번 발생할 수 있기 때문에 이 경우 기울기를 대체하는 대신 기울기를 합산해야 하기 때문입니다. 따라서 배치의 기울기를 계산하기 전에 optimizer.zero_grad()를 호출하는 것을 잊지 마세요.

# Training
마침내, 우리는 우리의 모델을 훈련시킬 준비가 되었습니다. 첫 번째 단계로, 우리는 약간 더 큰 데이터 세트를 만들고 더 큰 배치 크기를 가진 데이터 로더를 지정합니다.

```python
train_dataset = XORDataset(size=1000)
train_data_loader = data.DataLoader(train_dataset, batch_size=128, shuffle = True)
```

이제 우리는 작은 훈련 기능을 작성할 수 있습니다. 배치 로드, 예측 가져오기, 손실 계산, 역 전파 및 업데이트의 5단계를 기억하십시오. 또한 모든 데이터 및 모델 매개 변수를 선택한 장치(사용 가능한 경우 GPU)에 푸시해야 합니다. 우리가 가지고 있는 작은 신경망의 경우, 데이터를 GPU에 전달하는 것은 실제로 GPU에서 작업을 실행하는 것보다 훨씬 더 많은 시간이 걸립니다. 대규모 네트워크의 경우, 통신 시간이 실제 런타임보다 상당히 짧기 때문에 이러한 경우 GPU가 매우 중요합니다. 그럼에도 불구하고, 연습을 위해, 우리는 데이터를 여기서 GPU에 푸시할 것입니다.

```python
# 모델을 장치에 푸시합니다. 한 번만 수행해야 함
model.to(device)
```

또한 모델을 학습 모드로 설정합니다. 이는 model.train()을 호출하여 수행됩니다. 일부 모듈은 학습 중에 테스트 중에 수행하는 것과 다른 전방 단계가 필요합니다(예: BatchNorm 및 Dropout) 그리고 model.train() 및 model.eval()을 사용하여 이들 사이를 전환할 수 있습니다.

```python
def train_model(model, optimizer, data_loader, loss_module, num_epochs=100):
  # 모델을 train 모드로 설정
  model.train()

  #Trainig loop
  for epoch in tqdm(range(num_epochs)):
    for data_inputs, data_labels in data_loader:

      # 1단계: 입력 데이터를 장치로 이동(GPU를 사용하는 경우에만 엄격하게 필요)
      data_inputs = data_inputs.to(device)
      data_labels = data_labels.to(device)

      # 2단계: 입력 데이터에 대해 모델 실행
      preds = model(data_inputs)
      preds = preds.squeeze(dim=1) # 출력은 [배치 사이즈, 1]이지만 우리는 [배치 사이즈]를 원합니다

      # 3단계: 손실 계산
      loss = loss_module(preds, data_labels.float())

      # 4단계: 역전파 수행
      # 기울기를 계산하기 전에 기울기가 모두 0인지 확인해야 합니다.
      # 그라데이션은 덮어쓰지 않고 실제로 기존 그라데이션에 추가됩니다.
      optimizer.zero_grad()
      # 역전파 수행
      loss.backward()

      # 5단계: 매개 변수 업데이트
      optimizer.step()
```

```python
train_model(model, optimizer, train_data_loader, loss_module)
```

#Evalution
모델을 교육한 후에는 보류 테스트 세트에서 평가해야 합니다. 데이터 세트가 무작위로 생성된 데이터 포인트로 구성되어 있기 때문에 먼저 해당 데이터 로더로 테스트 세트를 만들어야 합니다.

```python
test_dataset = XORDataset(size=500)
# drop_last -> 마지막 배치가 128보다 작더라도 삭제하지 않음
test_data_loader = data.DataLoader(test_dataset, batch_size=128, shuffle=False, drop_last=False)
```

측정 기준으로 다음과 같이 계산된 정확도를 사용합니다:

image.png

여기서 TP는 참 양의 값, TN 참 음의 값, FP 거짓 양의 값 및 FN 오류 음의 값입니다.

모델을 평가할 때는 그레이디언트를 계산할 의도가 없기 때문에 계산 그래프를 추적할 필요가 없습니다. 이를 통해 필요한 메모리를 줄이고 모델의 속도를 높입니다. PyTorch에서 with,  torch.no _grad(): 를 사용하여 계산 그래프를 비활성화할 수 있습니다... 모델을 평가 모드로 추가 설정해야 합니다.

```python
def eval_model(model, data_loader):
  model.eval() # 모델을 평가 모드로 설정
  true_preds, num_preds = 0.0, 0.0

  with torch.no_grad(): # 다음 코드에 대한 그라데이션 비활성화
    for data_inputs, data_labels in data_loader:

      # 개발 집합에서 모형 예측 결정
      data_inputs, data_labels = data_inputs.to(device), data_labels.to(device)
      preds = model(data_inputs)
      preds = preds.squeeze(dim=1)
      preds = torch.sigmoid(preds) # 0과 1 사이의 예측을 매핑하는 시그모이드
      pred_labels = (preds >= 0.5).long() # 예측을 0과 1로 이항화

      # 정확도 메트릭에 대한 예측 기록 유지(true_proxys=TP+TN, num_dlls=TP+TN+FP+FN)
      true_preds += (pred_labels == data_labels).sum()
      num_preds += data_labels.shape[0]

  acc = true_preds / num_preds
  print(f"Accuracy of the model: {100.0*acc:4.2f}%")
```

```python
eval_model(model, test_data_loader)

#모델을 올바르게 교육했다면 100%에 가까운 정확도의 점수를 얻을 수 있습니다.
# 하지만, 이것은 우리의 간단한 과제 때문에 가능한 것이고, 불행히도, 우리는 보통 더 복잡한 과제의 시험 세트에서 그렇게 높은 점수를 받지 못합니다.
```

# Visualizing classfication boundaries (분류 경계 시각화)

모델이 학습한 내용을 시각화하기 위해 [-0.5, 1.5] 범위의 모든 데이터 포인트에 대해 예측을 수행하고 이 섹션의 시작 부분에 있는 샘플 그림에서와 같이 예측된 클래스를 시각화할 수 있습니다. 모형에서 결정 경계를 생성한 위치와 0으로 분류할 점, 1로 분류할 점이 표시됩니다. 따라서 파란색(클래스 0)과 주황색(클래스 1)에서 배경 이미지를 얻습니다. 모델이 불확실한 부분은 흐릿한 중첩을 볼 수 있습니다. 특정 코드는 클래스의 명확한 분리를 보여주는 출력 수치에 비해 관련성이 낮습니다:


```python
@torch.no_grad() # 데코레이터, 전체 기능에 대한 "with torch.no _grad(: ...)"와 동일한 효과.
def visualize_classification(model, data, label):
  if isinstance(data, Tensor):
    data = data.cpu().numpy()
  if isinstance(label, Tensor):
    label = label.cpu().numpy()
  data_0 = data[label == 0]
  data_1 = data[label == 1]

  plt.figure(figsize=(4,4))
  plt.scatter(data_0[:,0], data_0[:,1], edgecolor="#333",label="Class 0")
  plt.scatter(data_1[:,0], data_1[:,1], edgecolor="#333",label="Class 1")
  plt.title("Dataset samples")
  plt.ylabel(r"$x_2$")
  plt.xlabel(r"$x_1$")
  plt.legend()

  # 위에서 배운 많은 작업을 활용해 보겠습니다
  model.to(device)
  c0 = Tensor(to_rgba("C0")).to(device)
  c1 = Tensor(to_rgba("C1")).to(device)
  x1 = torch.arange(-0.5, 1.5, step=0.01, device=device)
  x2 = torch.arange(-0.5, 1.5, step=0.01, device=device)
  xx1, xx2 = torch.meshgrid(x1, x2) # 메쉬그리드 함수(numpy)
  model_inputs = torch.stack([xx1, xx2], dim=-1)
  preds = model(model_inputs)
  preds = torch.sigmoid(preds)

  # 차원에서 "없음"을 지정하면 새 차원이 생성됩니다
  output_image = (1 - preds)*c0[None, None] + preds*c1[None, None]
  output_image = ( output_image.cpu().numpy() )  # numpy 배열로 변환합니다. 이것은 CPU의 텐서에만 작동하므로 먼저 CPU에 푸시합니다
  plt.imshow(output_image, origin="lower", extent=(-0.5,1.5,-0.5,1.5))
  plt.grid(False)

visualize_classification(model, dataset.data, dataset.label)
plt.show()

#들여쓰기 조심
```

결정 경계는 CPU 또는 다른 GPU 아키텍처에서 실행할 때 발생할 수 있는 이 섹션의 서문에 있는 그림과 정확히 다르게 보일 수 있습니다. 그럼에도 불구하고 정확도 메트릭의 결과는 거의 동일해야 합니다.
